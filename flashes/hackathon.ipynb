```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.cvtColor(cv2.imread('defect1.png'),cv2.COLOR_BGR2GRAY)
im_res = cv2.resize(im,(int(im.shape[0]/2),int(im.shape[1]/2)))
im_comp = 255 - im_res
```

```python
plt.imshow(im_comp,cmap='gray')
plt.show()
```

```python
contours, hierarchy = cv2.findContours(im_comp, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
cnt = contours[1]
(x,y),radius = cv2.minEnclosingCircle(cnt)
center = np.array(int(x),int(y))
radius = int(radius)
```

```python
dist = []
for i in range(cnt.shape[0]):
    cord = cnt[i,:,:][0]
    radius = np.sqrt(np.sum(np.square(cord - center)))
    dist.append(radius)
```

```python
plt.plot(dist)
plt.show()
```
